```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator

def F(t,V):
    return np.array([-q*V[0]+q*V[1],-V[0]*V[2]+r*V[0]-V[1],V[0]*V[1]-b*V[2]])

#метод Рунге-Кутта 3-го порядка
def RK3(t0,T,y0,h,F):
    t_data = np.linspace(t0,T,int((T-t0)/h)+1)
    y_data = np.array([y0])
    for i in range(len(t_data)-1):
        k1 = F(t_data[i],y_data[i])
        k2 = F(t_data[i]+h/3,y_data[i]+h*k1/3)
        k3 = F(t_data[i]+2*h/3,y_data[i]+2*h*k2/3)
        y_cur = y_data[i] + h*(0.25*k1+0.75*k3)
        y_data = np.vstack([y_data,y_cur])
    return t_data,y_data
```

## Фазовые Портреты

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

#метод Рунге-Кутта 3-го порядка
def RK3(t0,T,y0,h,F):
    t_data = np.linspace(t0,T,int((T-t0)/h)+1)
    y_data = np.array([y0])
    for i in range(len(t_data)-1):
        k1 = F(t_data[i],y_data[i])
        k2 = F(t_data[i]+h/3,y_data[i]+h*k1/3)
        k3 = F(t_data[i]+2*h/3,y_data[i]+2*h*k2/3)
        y_cur = y_data[i] + h*(0.25*k1+0.75*k3)
        y_data = np.vstack([y_data,y_cur])
    return t_data,y_data

def PhasePortret(y_data):
    fig,axs = plt.subplots(1, 3, figsize = (15,5))
    X, Y = np.meshgrid(np.linspace(-5,5,y_data.shape[1]),np.linspace(-5,5,y_data.shape[1]))
    R, A = (X**2 + Y**2)**0.5, np.arctan2(Y,X)
    C, S = np.cos(A), np.sin(A)
    
    U, V = y_data[0] * C - R * S * y_data[1], y_data[0] * S + C * R * y_data[1] 
    axs[0].streamplot(X, Y, U, V, color = 'b', linewidth = 0.5, density = 1)
    axs[0].set_title("(x,y) proection")
    
    U, V = y_data[1] * C - R * S * y_data[2], y_data[1] * S + C * R * y_data[2]
    axs[1].streamplot(X, Y, U, V, color = 'b', linewidth = 0.5, density = 1)
    axs[1].set_title("(y,z) proection")
    
    U, V = y_data[0] * C - R * S * y_data[2], y_data[0] * S + C * R * y_data[2] 
    axs[2].streamplot(X, Y, U, V, color = 'b', linewidth = 0.5, density = 1)
    axs[2].set_title("(x,z) proection")
    


```

```python
h = 0.01
t0 = 0
T = 1
y0 = np.array([1,1,1])
q = 2
r = 4
b = 6
t_data1,y_data1 = RK3(t0,T,y0,h,F)
print(y_data1.shape)

fig,axs = plt.subplots(1,3,figsize =(15, 5))
axs[0].plot(t_data1,y_data1.transpose()[0], label ="Метод РК-3", color = "fuchsia", linewidth=2, zorder = 5)
axs[0].set_title('x(t)')

axs[1].plot(t_data1,y_data1.transpose()[1], label ="Метод РК-3", color = "fuchsia", linewidth=2, zorder = 5)
axs[1].set_title('y(t)')

axs[2].plot(t_data1,y_data1.transpose()[2], label ="Метод РК-3", color = "fuchsia", linewidth=2, zorder = 5)
axs[2].set_title('z(t)')

for i in range (3):
    axs[i].grid(which="major", linewidth=1.2)
    axs[i].grid(which="minor", linestyle="--", color="gray", linewidth=0.5)
    axs[i].tick_params(which='major', length=10, width=2)
    axs[i].tick_params(which='minor', length=5, width=1)
    axs[i].xaxis.set_minor_locator(AutoMinorLocator())
    axs[i].yaxis.set_minor_locator(AutoMinorLocator())
    axs[i].legend()

PhasePortret(y_data1.transpose())
```
